```python
import pandas as pd 
import torch 
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
import torch.nn as nn
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("enhanced_anxiety_dataset.csv")
df.head(3)
```

```python
df = df.drop(columns=['Therapy Sessions (per month)'])
```

```python
df.head(2)
```

```python
dummies = pd.get_dummies(df , columns=['Gender','Occupation' ,'Smoking' , 'Dizziness' ,'Medication' ,'Recent Major Life Event' ,'Family History of Anxiety' ]).astype(int)
```

```python
dummies.head(2)
```

```python
num_cols = df.drop(columns=['Gender','Occupation' ,'Smoking' , 'Dizziness' ,'Medication' ,'Recent Major Life Event' ,'Family History of Anxiety' ])
```

```python
df = pd.concat([dummies , num_cols] , axis=1)
```

```python
df.head(2)
```

```python
x= df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
scaled = StandardScaler()
x = scaled.fit_transform(x)
```

```python
x
```

```python
labled = LabelEncoder()
y = labled.fit_transform(y)
```

```python
y
```

```python
x = torch.from_numpy(x).float()
y = torch.from_numpy(y).float()
```

```python
import torch
import torch.nn as nn

class Model(nn.Module):  
    def __init__(self, num_feature):  
        super(Model, self).__init__()
        self.network = nn.Sequential(
            nn.Linear(num_feature, 5),
            nn.ReLU(),
            nn.Linear(5, 1),
            nn.Sigmoid()
        )

    def forward(self, features): 
        output = self.network(features)
        return output
```

```python
model = Model(x.shape[1])
```

```python
model.forward(x)
```
